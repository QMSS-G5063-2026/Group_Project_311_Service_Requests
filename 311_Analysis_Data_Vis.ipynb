```python
df["Created Date"] = pd.to_datetime(
    df["Created Date"],
    format="%m/%d/%Y %I:%M:%S %p",
    errors="coerce"
)

df["Closed Date"] = pd.to_datetime(
    df["Closed Date"],
    format="%m/%d/%Y %I:%M:%S %p",
    errors="coerce"
)
```

```python
df["Complaint Year"] = df["Created Date"].dt.year
```

```python
df["Complaint Month"] = df["Created Date"].dt.month
```

```python
print(df["Complaint Year"].unique())
df["Complaint Month"].unique()
```

```python
df["Month Name"] = pd.to_datetime(df["Complaint Month"], format="%m").dt.month_name()
```

```python
#for the ease of use - we will rename the columns
df = df.rename(columns={
    "Problem (formerly Complaint Type)": "Complaint",
    "Problem Detail (formerly Descriptor)": "Complaint Detail"
})
```

```python
df = df.rename(columns={
    'X Coordinate (State Plane)': "X Coordinate",
    'Y Coordinate (State Plane)': "Y Coordinate"
})
```

```python
grouping_complaints = (
    df.groupby(["Agency", "Complaint","Complaint Detail"])
      .size()
      .reset_index(name="count")
)
```
