## Mapping Neighborhoods & Fixing ZIP Codes

#### We have 524 missing ZIP codes and no neighborhood labels. Using the Latitude and Longitude coordinates, we will perform a spatial join with official NYC Shapefiles to:

#### Assign a Neighborhood (NTA) to every row so the HTML dashboard dropdown works.

#### Backfill missing ZIP codes where the data currently shows '00000'.

```python
!pip install geopandas
```

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point

# Loading the GeoJSON boundaries from the local files
nta_path = "2020_Neighborhood_Tabulation_Areas_(NTAs)_20260318.geojson"
zip_path = "Modified_Zip_Code_Tabulation_Areas_(MODZCTA)_20260318.geojson"

print("Loading local NYC Shapefiles...")
nta_gdf = gpd.read_file(nta_path)
zips_gdf = gpd.read_file(zip_path)

# Safety check for column names
if 'ntaname' in nta_gdf.columns:
    nta_gdf = nta_gdf.rename(columns={'ntaname': 'NTAName'})
if 'modzcta' in zips_gdf.columns:
    zips_gdf = zips_gdf.rename(columns={'modzcta': 'MODZCTA'})

# Preparing the 311 Data for Spatial Analysis
gdf_311 = gpd.GeoDataFrame(
    df, 
    geometry=gpd.points_from_xy(df.Longitude, df.Latitude),
    crs="EPSG:4326"
)

# Mapping to the Neighborhoods
print("Mapping coordinates to Neighborhood names...")
gdf_311 = gpd.sjoin(gdf_311, nta_gdf[['NTAName', 'geometry']], how="left", predicate="within")
gdf_311 = gdf_311.rename(columns={'NTAName': 'Neighborhood'})

# Dropping 'index_right' so the next join doesn't crash
if 'index_right' in gdf_311.columns:
    gdf_311 = gdf_311.drop(columns=['index_right'])

# Backfilling Missing ZIP Codes
missing_zip_mask = gdf_311['Incident Zip'].isin(['00000', '0', None, '', 'Unspecified'])
print(f"Backfilling {missing_zip_mask.sum()} missing ZIP codes using spatial join...")

# Only running the second join if there are actually missing ZIPs to find
if missing_zip_mask.any():
    zips_joined = gpd.sjoin(gdf_311[missing_zip_mask], zips_gdf[['MODZCTA', 'geometry']], how="left", predicate="within")
    
    # Updating the main 'Incident Zip' column
    gdf_311.loc[zips_joined.index, 'Incident Zip'] = zips_joined['MODZCTA']
    
    # Clean up index_right again after the second join
    if 'index_right' in gdf_311.columns:
        gdf_311 = gdf_311.drop(columns=['index_right'])

# Final Cleanup stuff
df = pd.DataFrame(gdf_311.drop(columns='geometry'))

print("\n--- Processing Complete ---")
print(f"Total Neighborhoods mapped: {df['Neighborhood'].notnull().sum()}")
print(f"Remaining '00000' ZIP codes: {(df['Incident Zip'] == '00000').sum()}")

# Saving the absolute final version
df.to_csv("NYC_311_Master_2024_2025.csv", index=False)
print("Master file saved! Ready for visualization.")
```

```python
import pandas as pd
import os

MASTER_FILE = "NYC_311_Master_2024_2025.csv"

def run_master_check(file_path):
    if not os.path.exists(file_path):
        print(f"ERROR: {file_path} not found.")
        return

    print(f"Auditing Master File: {file_path}\n" + "="*40)
    
    # Loading just a sample for speed
    df_check = pd.read_csv(file_path, nrows=100, low_memory=False)
    
    # Row Count Check 
    with open(file_path, 'rb') as f:
        row_count = sum(1 for line in f) - 1
    
    # Required Column Check
    required_cols = {
        'Geospatial (Map)': ['Latitude', 'Longitude', 'Incident Zip', 'Neighborhood'],
        'Analytics (Charts)': ['Created Date', 'Complaint', 'Borough'],
        'Word Cloud': ['Complaint Detail']
    }
    
    all_present = True
    for category, cols in required_cols.items():
        missing = [c for c in cols if c not in df_check.columns]
        if missing:
            print(f"{category} is missing: {missing}")
            all_present = False
        else:
            print(f"{category} columns verified.")

    # Data Integrity Sample
    print("-" * 40)
    print(f"Total Rows: {row_count:,}")
    
    # Checking if 'Created Date' is actually a date
    try:
        sample_date = pd.to_datetime(df_check['Created Date']).iloc[0]
        print(f"Date Format: Valid ({sample_date})")
    except:
        print("Warning: 'Created Date' needs conversion in data_loader.py")

    # Final Verdict
    print("=" * 40)
    if all_present and row_count > 5000000:
        print("VERDICT: MASTER FILE READY FOR DEPLOYMENT.")
    else:
        print("VERDICT: FILE NEEDS ALIGNMENT (See errors above).")

run_master_check(MASTER_FILE)
```
