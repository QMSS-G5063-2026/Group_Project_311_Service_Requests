## need to fix zipcode - make it str

```python
# Make sure Incident Zip is numeric (drop decimals) and then convert to string
df['Incident Zip'] = df['Incident Zip'].fillna(0).astype(int).astype(str)

# Optional: pad with leading zeros if needed (NYC zips are 5 digits)
df['Incident Zip'] = df['Incident Zip'].str.zfill(5)
```
